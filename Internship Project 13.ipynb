```python
df['Date_of_Journey']=pd.to_datetime(df['Date_of_Journey'])
df
```

```python
df['Year']=df['Date_of_Journey'].dt.year
df['Month']=df['Date_of_Journey'].dt.month
df['Day']=df['Date_of_Journey'].dt.day
df
```

```python
df['Arrival_Time']=pd.to_datetime(df['Arrival_Time'])
df
```

```python
df['Arrival_Hours']=df['Arrival_Time'].dt.hour
df['Arrival_Minutes']=df['Arrival_Time'].dt.minute
df
```

```python
def duration(x):
    x=x.split(' ')
    hours=0
    mins=0
    if len(x)==1:
        x=x[0]
        if x[-1]=='h':
            hours=int(x[:-1])
        else:
            mins=int(x[:-1])
    else:
        hours=int(x[0][:-1])
        mins=int(x[1][:-1])
    return hours,mins
df['Duration_hours']=df.Duration.apply(lambda x:duration(x)[0])
df['Duration_mins']=df.Duration.apply(lambda x:duration(x)[1])
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
maxAcc=0
maxRS=0
for i in range(1,100):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.32,random_state=i)
    mod=DecisionTreeRegressor()
    mod.fit(x_train,y_train)
    pred=mod.predict(x_test)
    score=mod.score(x_train,y_train)
    if score>maxAcc:
        maxAcc=score
        maxRS=i
print('The best accuracy is',maxAcc,'on random state',maxRS)        
```

```python
lm=LinearRegression()
lm.fit(x_train,y_train)
lm.score(x_train,y_train)
```

```python
pred=lm.predict(x_test)
print('Predicted price is',pred)
print('Actual price is',y_test)
```

```python
print(mean_squared_error(y_test,pred))
print(mean_absolute_error(y_test,pred))
print('Root mean squared error',np.sqrt(mean_squared_error(y_test,pred)))
print(r2_score(y_test,pred))
```

```python
dtr=DecisionTreeRegressor()
dtr.fit(x_train,y_train)
dtr.score(x_train,y_train)
```

```python
pred=dtr.predict(x_test)
print('Predicted price is',pred)
print('Actual price is',y_test)
```

```python
print(mean_squared_error(y_test,pred))
print(mean_absolute_error(y_test,pred))
print('Root mean squared error',np.sqrt(mean_squared_error(y_test,pred)))
print(r2_score(y_test,pred))
```

```python
knn=KNeighborsRegressor(n_neighbors=3)
knn.fit(x_train,y_train)
knn.score(x_train,y_train)
```

```python
pred=knn.predict(x_test)
print('Predicted price is',pred)
print('Actual price is',y_test)
```

```python
print(mean_squared_error(y_test,pred))
print(mean_absolute_error(y_test,pred))
print('Root mean squared error',np.sqrt(mean_squared_error(y_test,pred)))
print(r2_score(y_test,pred))
```

```python
rf=RandomForestRegressor()
rf.fit(x_train,y_train)
rf.score(x_train,y_train)
```

```python
pred=rf.predict(x_test)
print('Predicted price is',pred)
print('Actual price is',y_test)
```

```python
print(mean_squared_error(y_test,pred))
print(mean_absolute_error(y_test,pred))
print('Root mean squared error',np.sqrt(mean_squared_error(y_test,pred)))
print(r2_score(y_test,pred))
```

These scores can be because of overfitting. Lets check with cross validation.

# Cross Validation

```python
from sklearn.model_selection import cross_val_score
score=cross_val_score(lm,x,y,cv=5)
print(score)
print(score.mean())
```

```python
score=cross_val_score(dtr,x,y,cv=5)
print(score)
print(score.mean())
```

```python
score=cross_val_score(knn,x,y,cv=5)
print(score)
print(score.mean())
```

```python
score=cross_val_score(rf,x,y,cv=5)
print(score)
print(score.mean())
```

The least difference between the r2_score and cv score says that decision tree regressor model is working better.
